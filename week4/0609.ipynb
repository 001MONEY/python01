# Git Bash 基本命令总结

## 配置相关
| 命令 | 说明 |
|------|------|
| `git config --global user.name "姓名"` | 设置全局用户名 |
| `git config --global user.email "邮箱"` | 设置全局邮箱 |
| `git config --list` | 查看当前配置 |

## 仓库初始化与克隆
| 命令 | 说明 |
|------|------|
| `git init` | 在当前目录初始化 Git 仓库 |
| `git clone <仓库地址>` | 克隆远程仓库到本地 |

## 基本操作
| 命令 | 说明 |
|------|------|
| `git status` | 查看工作区状态 |
| `git add <文件>` | 将文件添加到暂存区 |
| `git add .` | 添加所有修改到暂存区 |
| `git commit -m "提交信息"` | 提交暂存区内容到本地仓库 |
| `git commit -a -m "提交信息"` | 跳过暂存区，直接提交已跟踪的修改 |
| `git log` | 查看提交历史 |
| `git log --oneline` | 以简洁方式查看提交历史 |
| `git diff` | 查看工作区与暂存区的差异 |
| `git diff --cached` | 查看暂存区与最新提交的差异 |

## 分支管理
| 命令 | 说明 |
|------|------|
| `git branch` | 查看本地分支列表 |
| `git branch <分支名>` | 创建新分支 |
| `git checkout <分支名>` | 切换到指定分支 |
| `git checkout -b <分支名>` | 创建并切换到新分支 |
| `git merge <分支名>` | 将指定分支合并到当前分支 |
| `git branch -d <分支名>` | 删除本地分支 |

## 远程仓库
| 命令 | 说明 |
|------|------|
| `git remote -v` | 查看远程仓库地址 |
| `git push` | 推送本地提交到远程仓库 |
| `git push origin <分支名>` | 推送指定分支到远程 |
| `git pull` | 拉取远程最新代码并合并 |
| `git fetch` | 拉取远程最新代码但不合并 |

## 撤销与回退
| 命令 | 说明 |
|------|------|
| `git restore <文件>` | 撤销工作区的修改 |
| `git restore --staged <文件>` | 将文件从暂存区移出 |
| `git reset --soft HEAD~1` | 撤销最近一次提交，保留修改 |
| `git reset --hard HEAD~1` | 撤销最近一次提交，丢弃修改 |
| `git revert <提交ID>` | 通过新提交撤销指定提交的更改 |

## 其他实用命令
| 命令 | 说明 |
|------|------|
| `git stash` | 暂存当前工作区的修改 |
| `git stash pop` | 恢复最近一次暂存的修改 |
| `git tag` | 查看标签列表 |
| `git tag <标签名>` | 创建标签 |
| `git rm <文件>` | 删除文件并加入暂存区 |
| `git mv <旧名> <新名>` | 重命名文件并加入暂存区 |
