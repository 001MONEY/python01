```python
# 大写字母 A-Z
print("大写字母 A-Z:")
i = ord('A')
while i <= ord('Z'):
    print(f"{chr(i)} → {i}")
    i += 1

print("\n------------------------\n")

# 小写字母 a-z
print("小写字母 a-z:")
i = ord('a')
while i <= ord('z'):
    print(f"{chr(i)} → {i}")
    i += 1
```

作业1 求解1-100中能被3整除的所有数字的和

```python
s = 0
for i in range(1,101):
    if i % 3 == 0:
        s += i
print(s)
```

作业2求解1-100 能被3和5同时整除的数字的和

```python
sum_i = 0
for i in range(1,101):
    if i % 3 == 0 and i % 5 == 0:
        sum_i += i
print(sum_i)
```

作业5输入一个数字将其转换为二进制，判断二进制中1的个数

```python
num = int(input("请输入一个数字："))
count = 0
n = num  
while n > 0:
    # 取n除以2的余数，如果是1，就计数+1
    if n % 2 == 1:
        count += 1
    # 把n整除2，相当于右移一位
    n = n // 2

print(f"数字 {num} 的二进制中,1的个数是:{count}")
```
