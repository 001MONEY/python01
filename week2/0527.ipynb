函数基础：声明、调用、返回值

先声明再调用——Python 从上往下扫描函数定义，声明必须在调用前面。日常写法：前面全部声明函数，末尾统一调用。

return 做两件事：把结果送出去，同时终止函数。函数里遇到 return，后面的代码全不执行。没写 return 或只写了空 return，返回 None。

多个返回值：一个 return 后面跟多个值，Python 自动打包成元组。用对应数量的变量接收，就拆开成各自类型；用 1 个变量接收，就是元组。

匿名函数 lambda：没有名字，冒号前是形参，冒号后是返回值，自带 return。只能写简单逻辑，复杂判断和循环还是得用 def。

pass 占位符：先搭结构再填逻辑，保证程序能跑。

```python
#定义 一个函数来计算两个数的和
def sum(a, b):
    sum = a + b  #计算 a 和 b 的和
    return sum  #返回 a 和 b 的和
def main():
    print(sum(3, 5))#调用 sum 函数并打印结果
if __name__ == "__main__":
    main()
```


| 参数类型  | 声明方式                 | 调用规则              |
| ----- | -------------------- | ----------------- |
| 必须参数  | `def f(a, b)`        | 顺序传，缺了报 TypeError |
| 关键字参数 | 调用时 `f(age=16)`      | 一旦出现，后面全得用关键字     |
| 默认参数  | `def f(a, h=1.88)`   | 传了用传入值，没传用默认值     |
| 不定长参数 | `*args` / `**kwargs` | 一颗星打包成元组，两颗星打包成字典 |
打包 vs 解包：声明时用星号叫打包（多个实参 → 元组/字典），调用时用星号叫解包（元组/字典 → 多个实参）。元组一颗星，字典两颗星。
位置陷阱：不定长参数放前面会"吃掉"所有实参，后面的必须参数只能用关键字参数传。

```python
# 假设这是你定义的函数
def total_price(cut_off, *prices):
    total = sum(prices)
    if total > cut_off:
        return total - cut_off
    else:
        return total
# 准备数据
prices_list = [10, 20, 30, 15, 25]   # 总价为 100
cut_off = 65  # 假设满减门槛是 65

# ✅ 正确的调用方式：
# 1. cut_off 使用关键字参数传值
# 2. 列表前加 * 进行解包，且不能加等号
ret = total_price(cut_off, *prices_list)  # ✅ 正确的调用方式 

print(ret)  # 输出 35（因为总价 100 > 65，所以返回 100 - 65 = 35）
```

嵌套函数与闭包

嵌套函数：函数里再定义函数。内部函数只能在外部函数内部调用，外面找不到。

作用域方向：内部可以访问外部的变量，外部不能访问内部的变量——单向可见。

闭包三个条件：①有嵌套函数 ②内部函数引用外部函数的变量 ③外部函数返回内部函数名（不加括号）。

闭包 vs 普通函数：普通函数执行完变量就释放了；闭包能保持外部变量状态，只要内部函数对象没被回收，外部变量就一直被引用。

入门阶段定位：能看懂闭包结构就行，不需要自己写。"函数可以携带状态"这个认知要留住——装饰器、回调都是它的一脉相承。

```python
#嵌套函数
def outer_func(x):
    def inner_func(y):
        return x + y
    return inner_func
add_eight = outer_func(8)  # 创建一个函数，固定 x 为 8
result = add_eight(10)  # 调用内层函数，传入 y 为 10
print(result)  # 输出 18，因为 8 + 10 = 18  
```

```python
#闭包示例
def make_multiplier(n):
    def multiplier(x):
        return x * n
    return multiplier
times3 = make_multiplier(3)  # 创建一个函数，固定 n 为 3
result = times3(10)  # 调用内层函数，传入 x 为 10   
print(result)  # 输出 30，因为 10 * 3 = 30
```

冒泡排序：第一个算法
核心规则：相邻元素两两比较，前面比后面大就交换，每轮把最大值"冒泡"到最右。
手动推导（5个元素）：
- 第1轮：比较4次，确定最大值在最右
- 第2轮：比较3次，确定第二大值
- 第3轮：比较2次
- 第4轮：比较1次，排序完成
关键理解：外层控制行（轮次），内层控制列（每轮比较）。内层循环次数 = 列表长度 - 1 - 外层计数器。理解不了减 j 也没关系，不减只是多几次无效比较，不影响结果。
选择排序（Selection Sort）是一种简单直观的排序算法。它的核心思想非常符合人类的直觉：每一轮都从剩下的未排序元素中“挑选”出最小（或最大）的一个，放到已排序序列的末尾。

```python
#冒泡排序
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):                                  # 外层循环控制所有的遍历次数
        for j in range(0, n-i-1):                       # 内层循环控制每次遍历的范围，逐渐减少
            if arr[j] > arr[j+1]:                       # 如果前一个元素大于后一个元素，则交换它们
                arr[j], arr[j+1] = arr[j+1], arr[j]     # 交换元素
    return arr                                          # 返回排序后的数组
arr = [64, 34, 25, 12, 22, 11, 90]
sorted_arr = bubble_sort(arr)                           # 调用冒泡排序函数
print(sorted_arr)                                       # 输出 [11, 12, 22, 25, 34, 64, 90]     

#选择排序
def selection_sort(arr):
    n = len(arr)
    for i in range(n):                              # 外层循环控制所有的遍历次数
        min_idx = i                                 # 假设当前元素是最小的
        for j in range(i+1, n):                     # 内层循环从当前元素的下一个位置开始，寻找更小的元素
            if arr[j] < arr[min_idx]:               # 如果找到更小的元素，更新最小元素的索引
                min_idx = j                         # 交换当前元素和最小元素
        arr[i], arr[min_idx] = arr[min_idx], arr[i] # 交换元素
    return arr                                      # 返回排序后的数组
arr = [64, 34, 25, 12, 22, 11, 90]
sorted_arr = selection_sort(arr)
print(sorted_arr)  # 输出 [11, 12, 22, 25, 34, 64, 90]

#快速排序
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less_than_pivot = [x for x in arr[1:] if x <= pivot]
        greater_than_pivot = [x for x in arr[1:] if x > pivot]
        return quick_sort(less_than_pivot) + [pivot] + quick_sort(greater_than_pivot)
arr = [10, 7, 8, 9, 1, 5]
sorted_arr = quick_sort(arr)
print(sorted_arr)  # 输出 [1, 5, 7, 8, 9, 10]
```

```python
#作业4给定列表 nums = [10, 20, 30, 50, 20], 定义一个函数找出给定元素的所有位置
def find_all_positions(nums, target):
    positions = []  # 用于存储目标元素的所有位置
    for index, num in enumerate(nums):  # 使用 enumerate 获取元素及其索引
        if num == target:  # 如果当前元素等于目标元素
            positions.append(index)  # 将索引添加到位置列表中
    return positions  # 返回所有位置的列表
nums = [10, 20, 30, 50, 20]
target = 20
positions = find_all_positions(nums, target)  # 调用函数并获取结果
print(positions)  # 输出 [1, 4]，因为 20 出现在索引 1 和 4 的位置
```

```python
#作业6定义一个函数,实现返回一个数的阶乘
# 尾递归版本
def factorial_tail(n, acc=1):
    if n <= 1:
        return acc  # 👈 达到基准条件，直接返回最终结果
    return factorial_tail(n - 1, n * acc)  # 👈 递归调用是函数的最后一步操作，没有任何后续计算
result = factorial_tail(5)  # 调用函数并获取结果
print(result)  # 输出 120，因为 5! = 5 * 4 * 3 * 2 * 1 = 120
```

```python
#作业8求前 K 个高频元素设计对应的函数
from collections import Counter
def top_k_frequent(nums, k):
    count = Counter(nums)  # 统计每个元素的频率
    most_common = count.most_common(k)  # 获取前 K 个最常见的元素及其频率
    return [item[0] for item in most_common]  # 返回前 K 个高频元素的列表
nums7 = [2, 3, 1, 2, 3, 3, 3, 4, 4, 4]
k = 2
result = top_k_frequent(nums7, k)  # 调用函数并获取结果
print(result)  # 输出 [3, 4]，因为 3 出现了4次，4 出现了3次，是前 2 个高频元素  
```

```python
def fun8(nums, k):
    # 1. 统计每个元素出现的次数
    dic = {}
    for num in nums:
        dic[num] = dic.get(num, 0) + 1

    # 2. 字典转换为列表 [(数字, 次数), ...]
    lst_nums = []
    for key, v in dic.items():
        lst_nums.append((key, v))

    # 3. 部分选择排序（只排前 k 个，降序）
    for j in range(k):
        max_idx = j
        for i in range(j + 1, len(lst_nums)):
            if lst_nums[max_idx][1] < lst_nums[i][1]:
                max_idx = i
        # 将当前最大值交换到位置 j
        lst_nums[j], lst_nums[max_idx] = lst_nums[max_idx], lst_nums[j]

    # 4. 提取前 k 个高频元素
    result = []
    for num_tuple in lst_nums[:k]:
        result.append(num_tuple[0])

    return result


# 测试
nums7 = [2, 3, 1, 2, 3, 3, 3, 4, 4, 4]
print(f"前 K 个高频元素: {fun8(nums7, k)}")
```

```python
def fun8(nums, k):
    """
    求列表中前 k 个高频元素（纯基础语法实现版）
    :param nums: 输入的整数列表
    :param k: 需要返回的高频元素个数
    :return: 包含前 k 个高频元素的列表
    """
    # ==================== 第1步：手动统计词频 ====================
    dic = {}                              # 创建一个空字典，用来存放 {数字: 出现次数}
    for num in nums:                      # 遍历输入列表中的每一个数字
        dic[num] = dic.get(num, 0) + 1    # 用 get 方法安全取值：如果字典里已有该数字，就在原次数上+1；如果没有，get返回默认值0，0+1=1完成首次记录

    # ==================== 第2步：字典转列表 ====================
    lst_nums = []                         # 创建一个空列表，用来存放 (数字, 次数) 的元组对
    for key, v in dic.items():            # 遍历字典的所有键值对，key是数字，v是对应的出现次数
        lst_nums.append((key, v))         # 将数字和次数打包成元组，追加到列表中（因为字典无法直接按值排序，必须转成列表）

    # ==================== 第3步：部分选择排序（降序） ====================
    for j in range(k):                    # 外层循环只执行 k 次，因为我们只需要找出前 k 个最大值，无需对整个列表全排序
        max_idx = j                       # 假设当前位置 j 就是剩余未排序部分中最大值的索引
        for i in range(j + 1, len(lst_nums)):  # 内层循环从 j+1 开始，遍历剩余所有元素，寻找真正的最大值
            if lst_nums[max_idx][1] < lst_nums[i][1]:  # 比较次数：[1]代表取元组的第二个元素（即出现次数），如果当前最大值次数小于i位置的次数
                max_idx = i               # 更新最大值索引为 i，说明找到了更大的频次
        # 内层循环结束后，max_idx 指向的就是剩余部分中频次最高的元素
        lst_nums[j], lst_nums[max_idx] = lst_nums[max_idx], lst_nums[j]  # Python特有的交换语法：把找到的最大值换到位置j，完成本轮选择

    # ==================== 第4步：提取结果 ====================
    result = []                           # 创建空列表，用来存放最终要返回的数字
    for num_tuple in lst_nums[:k]:        # 切片取出排好序的前 k 个元组，逐个遍历
        result.append(num_tuple[0])       # [0]代表取元组的第一个元素（即数字本身），追加到结果列表中

    return result                         # 返回最终结果


# ==================== 测试代码 ====================
nums7 = [2, 3, 1, 2, 3, 3, 3, 4, 4, 4]  # 定义测试数据
print(f"前 K 个高频元素: {fun8(nums7, 3)}")  # 调用函数并格式化输出结果
```
